## 課題1

各問で示されている図を出力してください。

その際、以下の要件を必ず満たすようにしてください。


1.   for文を必ず用いること
2.   変数`size`の値を変えることで自由に図形の大きさを変更できること




### 問1

以下の図形を出力してください。

```
*
**
***
****
*****
```

````python
size = 5
for i in range(1, size + 1):
    print('*' * i)
````

### 問2

以下の図形を出力してください。

```
*
 *
  *
   *
    *
```

````python
size = 5
for i in range(size):
    print(' ' * i + '*')
````

### 問3

以下の図形を出力してください。

```
    *
   *
  *
 *
*
```

````python
size = 5
for i in range(size):
    print(' ' * (size - i - 1) + '*')
````

### 問4

以下の図形を出力してください。

```
*   *
 * *
  *
 * *
*   *
```

````python
size = 5
for i in range(size):
    if i < size // 2:
        print(' ' * i + '*' + ' ' * (size - 2 - 2 * i) + '*')
    elif i == size // 2:
        print(' ' * i + '*')
    else:
        print(' ' * (size - i - 1) + '*' + ' ' * (2 * (i - size // 2) - 1) + '*')
````

### 問5

以下の図形を出力してください。

```
    *
   ***
  *****
 *******
*********
```

````python
size = 5
for i in range(size):
    print(' ' * (size - i - 1) + '*' * (2 * i + 1))
````
