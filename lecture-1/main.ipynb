```python
import time
import requests
from bs4 import BeautifulSoup

repo_list_url = "https://github.com/orgs/google/repositories"
```

```python
# result_list = []
response = requests.get(repo_list_url)
# result_list.append(response.text)
response.text
```

```python
# bs4 init
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Find the maximum number of pages
max_pages = soup.find("nav", {"aria-label": "Pagination"}).find_all("a")[-2].text
max_pages
```

```python
# get all pages
result_list = []
for i in range(int(max_pages)):
    this = f"{repo_list_url}?page={i+1}"
    response = requests.get(this)
    result_list.append(response.text)
    print(f"Fetched page {i+1}: {response.text[:50].strip()}")
    time.sleep(1)
print(len(result_list))
```

```python

import re
repos_list_parser = re.compile("ListView-module__ul")
repo_title_parser = re.compile("Title-module__anchor")
repo_using_langs_parser = re.compile("ReposListItem-module__Text_4")
repo_stars_parser = re.compile("ReposListItem-module__Link_1") # need index

parsed_results = []
for index_, page in enumerate(result_list):
    this_page = BeautifulSoup(page, 'html.parser')
    repos_list = this_page.find("ul", {"class": repos_list_parser}).find_all("li")
    len(repos_list)
    # parsed_result_of_this_page = []
    for index2_, i in enumerate(repos_list):
        print(f"{index_+1}/{index2_+1}")
        title = i.find("a", {"class": repo_title_parser}).find("span").text
        try:
            lang = i.find("span", {"class": repo_used_langs_parser}).text
        except AttributeError:
            lang = "N/A"
        # print(len(i.find_all("a", {"class": repo_stars_parser})))
        stars = i.find_all("a", {"class": repo_stars_parser})[1].text
        # parsed_result_of_this_page.append([title, lang, stars])
        parsed_results.append([title, lang, stars])
    # parsed_results.extend(parsed_result_of_this_page)
```

```python
print("\n".join(str(_) for _ in parsed_results))
```

```python
# create db and table
import sqlite3
path = ''
db_name = 'google_repos.db'
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = 'CREATE TABLE repos (id INTEGER PRIMARY KEY AUTOINCREMENT, title TEXT, lang TEXT, stars TEXT);'
    cur.execute(sql)
except sqlite3.Error as e:
    print('エラーが発生しました:', e)
finally:
    conn.close()
```

```python
# insert data
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    for repo in parsed_results:
        title, lang, stars = repo
        sql = f"INSERT INTO repos (title, lang, stars) VALUES ('{title}', '{lang}', '{stars}');"
        cur.execute(sql)
    conn.commit()
except sqlite3.Error as e:
    print('エラーが発生しました:', e)
finally:
    conn.close()
```

```python
# print data from db
try:
    conn = sqlite3.connect(path + db_name)
    cur = conn.cursor()
    sql = "SELECT * FROM repos;"
    cur.execute(sql)
    rows = cur.fetchall()
    for row in rows:
        # print(row)
        print(f"ID: {row[0]}, Title: {row[1]}, Language: {row[2]}, Stars: {row[3]}")
except sqlite3.Error as e:
    print('エラーが発生しました:', e)
finally:
    conn.close()
```
